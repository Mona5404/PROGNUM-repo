```python
#Session quiz 4.4
import numpy as np
from matplotlib import pyplot as plt
import scipy

ppl = np.random.normal(185, 8, 100000) #Random gaussian sample, loc = 185 stdev = 8, size = 100000
plt.hist(ppl, bins=40) #make a histogram with 40 bins with the sample

#Average and stdev calculation
mu = np.sum(ppl) / np.size(ppl)
stdev = np.std(ppl)
print(f"The mean is {mu} and the standard deviation is {stdev}")

#Plot (further)
plt.axvline(x=mu, color='green')
up = mu - stdev
low = mu + stdev

#Shade 1 stdev of area
plt.axvspan(low, up, alpha=0.4, color='pink')

#norm.cdf
h = float(input(f"Height:"))#Height input
a = scipy.stats.norm.cdf(h, loc=185, scale=8) #Probability that random number X is smaller than or equal to the height input
#Probability that height is within 1 standard deviation range 
b = scipy.stats.norm.cdf(up,loc=185, scale=8) 
c = scipy.stats.norm.cdf(low,loc=185, scale=8)
d = c - b

print(f"The probability that a random value is smaller than or equal to that value is {a}")
print(f"The probability of a height located within 1 standard deviation range is {d}")
```

```python
#SQ 4.6 
import numpy as np 
from numpy import *

#Input the function 
func = input(f"Function:")
a = float(input(f"Lower bound (a):"))
b = float(input(f"Upper bound (b):"))
n = int(input(f"How many random values are to be used in the summation?:"))

#Define the Monte Carlo integration 
def mci(func, a, b, n):
    rndm = np.random.uniform(a, b, n)
    mci_1 = (b - a)/n
    mci_2 = eval(func, mapping=rndm)
    mci_3 = mci_1 * mci_2
    return mci_3

ans = mci(func, a, b, n) 
print(f"{ans}")
```

```python
#SQ 4.21
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Seed for reproducibility (optional)
np.random.seed(0)

# Generate data
N = 100000
X = np.random.normal(loc=0, scale=1, size=N)
Y = np.random.normal(loc=2, scale=1, size=N)

# 1D Histogram + PDF Plot
fig, ax = plt.subplots(figsize=(7, 6))
bins = 30

# Set range and bins manually for control of axis
hist_range = (-6, 8)
edges = np.linspace(hist_range[0], hist_range[1], bins + 1)

# Plot histogram for X (blue)
counts_X, bin_edges_X, patches_X = ax.hist(
    X,
    bins=edges,
    density=True,
    alpha=0.5,
    color="blue",
    label="X",
    edgecolor="black",
)

# Plot histogram for Y (red)
counts_Y, bin_edges_Y, patches_Y = ax.hist(
    Y,
    bins=edges,
    density=True,
    alpha=0.5,
    color="red",
    label="Y",
    edgecolor="black",
)

# Ensure histogram bar outlines are solid and of standard thickness
for p in patches_X + patches_Y:
    p.set_linewidth(1.2)
    p.set_linestyle("solid")

# Plot PDF for X and Y
x_pdf = np.linspace(hist_range[0], hist_range[1], 500)
ax.plot(
    x_pdf,
    norm.pdf(x_pdf, 0, 1),
    "k:",
    lw=3,
    label="X(PDF)",
)
ax.plot(
    x_pdf,
    norm.pdf(x_pdf, 2, 1),
    "k--",
    lw=3,
    label="Y(PDF)",
)

# Configure axes limits and ticks
ax.set_xlim(hist_range)
ax.set_ylim(0, 0.40)
ax.set_xticks(np.arange(hist_range[0], hist_range[1] + 1, 1))
ax.set_yticks(np.arange(0, 0.41, 0.05))

# Labels, title, grid, legend
ax.set_xlabel("Values")
ax.set_ylabel("Probability Density")
ax.set_title("1D Histogram + PDF")
ax.grid(axis="y", linestyle="--", alpha=0.6)
ax.legend()

# Annotations for distributions
ax.annotate(
    r"$\mathcal{N}(0,1)$",
    xy=(-1.1, 0.16),
    xytext=(-3.0, 0.19),
    arrowprops=dict(arrowstyle="->", color="blue", lw=2.2),
    color="blue",
    fontsize=12,
    ha="center",
)
ax.annotate(
    r"$\mathcal{N}(2,1)$",
    xy=(2.4, 0.16),
    xytext=(4.2, 0.19),
    arrowprops=dict(arrowstyle="->", color="red", lw=2.2),
    color="red",
    fontsize=12,
    ha="center",
)

plt.tight_layout()

# 2D Histogram + Contour Plot
fig2, ax2 = plt.subplots(figsize=(8, 6))

# 2D histogram (counts) with blue colormap
hist, xedges, yedges, img = ax2.hist2d(X, Y, bins=30, cmap="Blues")
cb = plt.colorbar(img, ax=ax2)

# Build grid of bin centers for contours
x_centers = 0.5 * (xedges[:-1] + xedges[1:])
y_centers = 0.5 * (yedges[:-1] + yedges[1:])
X_mesh, Y_mesh = np.meshgrid(x_centers, y_centers, indexing="ij")

# Use the histogram itself (from the data) as the "density" surface
Z = hist

# Compute contour levels so regions above each level contain
# approximately 68%, 95%, and 99.7% of the total counts.
Z_flat = Z.ravel()
Z_sorted = np.sort(Z_flat)[::-1]  # descending counts
cumsum = np.cumsum(Z_sorted) / Z_sorted.sum()

prob_levels = [0.6827, 0.9545, 0.9973]
raw_contour_levels = []
for p in prob_levels:
    idx = np.searchsorted(cumsum, p)
    idx = min(idx, len(Z_sorted) - 1)
    raw_contour_levels.append(Z_sorted[idx])

# Pair with labels so we can sort by increasing level (required by contour)
sigma_text = ["1σ", "2σ", "3σ"]
level_label_pairs = sorted(
    zip(raw_contour_levels, sigma_text), key=lambda t: t[0]
)
contour_levels = [lvl for lvl, _ in level_label_pairs]

# Red contour lines
CS = ax2.contour(
    X_mesh,
    Y_mesh,
    Z,
    levels=contour_levels,
    colors="red",
    linewidths=2,
)

# Label contours with 1σ, 2σ, 3σ directly on the plot
fmt = {level: label for level, label in level_label_pairs}
ax2.clabel(CS, CS.levels, inline=True, fmt=fmt, fontsize=10, colors="red")

ax2.set_title("2D Histogram + Contour")
ax2.set_xlabel("X")
ax2.set_ylabel("Y")
ax2.grid(True, linestyle="--", alpha=0.6)
ax2.set_aspect("equal", adjustable="box")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits

# Open the FITS file
hdul = fits.open('APOGEE_stars.fits')

# Access the primary data
data = hdul[1].data
colnames = hdul[1].columns.names  # DO NOT remove, per instructions

# Apply quality cuts
mask = (
    (data['STARFLAG'] == 0) &
    (data['ASPCAPFLAG'] == 0) &
    (data['SNR'] > 10) &
    (data['GAIAEDR3_PARALLAX'] / data['GAIAEDR3_PARALLAX_ERROR'] > 5) &
    (~np.isnan(data['GAIAEDR3_PHOT_G_MEAN_MAG'])) &
    (~np.isnan(data['GAIAEDR3_PHOT_BP_MEAN_MAG'])) &
    (~np.isnan(data['GAIAEDR3_PHOT_RP_MEAN_MAG']))
)

good_data = data[mask]

# Choose a number of points that best matches the image aspect
np.random.seed(42)
Nplot = 850
if len(good_data) > Nplot:
    plot_indices = np.random.choice(len(good_data), Nplot, replace=False)
    plot_data = good_data[plot_indices]
else:
    plot_data = good_data

# Extract HRD columns
parallax = plot_data['GAIAEDR3_PARALLAX']
m_G = plot_data['GAIAEDR3_PHOT_G_MEAN_MAG']
m_BP = plot_data['GAIAEDR3_PHOT_BP_MEAN_MAG']
m_RP = plot_data['GAIAEDR3_PHOT_RP_MEAN_MAG']
bp_rp = m_BP - m_RP
M_G = m_G + 5 * np.log10(parallax / 1000) + 5

# Define classes
is_giant = (M_G < 3.5) & (bp_rp > 0.8)
is_ms = ~is_giant

# Plotting
fig, ax = plt.subplots(figsize=(8,6))

# Dots: sizes and alpha tweaked for accuracy; edgecolor='none' for clean look
ax.scatter(bp_rp[is_ms], M_G[is_ms], s=10, c='blue', label='MS', alpha=0.84, edgecolor='none', linewidths=0)
ax.scatter(bp_rp[is_giant], M_G[is_giant], s=13, c='red', label='Giant Branch', alpha=0.95, edgecolor='none', linewidths=0)

# WD box as in the image
wd_box = plt.Rectangle((0, 10), 0.8, 6, color='gray', alpha=0.4, zorder=0, label="White Dwarf Region")
ax.add_patch(wd_box)
ax.text(0.18, 13.3, "WD", fontsize=16, color='black', 
        bbox=dict(facecolor='none', edgecolor='none', alpha=0.0), zorder=10)

# Thicker, shorter arrows with improved placement to match the image
ax.annotate('Giant Branch',
    xy=(1.33, 1.0), xycoords='data',
    xytext=(1.90, 0.35), textcoords='data',
    fontsize=13, color='red',
    arrowprops=dict(facecolor='red', edgecolor='red', shrink=0.07, width=3, headwidth=10, alpha=0.65, lw=1.45),
    zorder=11
)
ax.annotate('MS',
    xy=(1.0, 4.0), xycoords='data',
    xytext=(1.55, 4.70), textcoords='data',
    fontsize=13, color='blue',
    arrowprops=dict(facecolor='blue', edgecolor='blue', shrink=0.08, width=3, headwidth=10, alpha=0.70, lw=1.45),
    zorder=11
)

ax.set_xlabel(r"$BP - RP$", fontsize=15)
ax.set_ylabel(r"$M_G$", fontsize=15)
ax.set_title("HR Diagram", fontsize=17)

# Axes match the image
ax.set_xlim(0, 4)
ax.set_xticks(np.arange(0, 5, 1))
ax.set_ylim(15, -2)

# Legend: below the "Giant Branch" label area, not in the upper right
ax.legend(loc='upper center', bbox_to_anchor=(0.53, 0.46), frameon=True)
plt.tight_layout()
plt.show()

```
