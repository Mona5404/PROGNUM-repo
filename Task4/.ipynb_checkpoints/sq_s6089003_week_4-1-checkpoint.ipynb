```python
#Session quiz 4.4
import numpy as np
from matplotlib import pyplot as plt
import scipy

ppl = np.random.normal(185, 8, 100000) #Random gaussian sample, loc = 185 stdev = 8, size = 100000
plt.hist(ppl, bins=40) #make a histogram with 40 bins with the sample

#Average and stdev calculation
mu = np.sum(ppl) / np.size(ppl)
stdev = np.std(ppl)
print(f"The mean is {mu} and the standard deviation is {stdev}")

#Plot (further)
plt.axvline(x=mu, color='green')
up = mu - stdev
low = mu + stdev

#Shade 1 stdev of area
plt.axvspan(low, up, alpha=0.4, color='pink')

#norm.cdf
h = float(input(f"Height:"))#Height input
a = scipy.stats.norm.cdf(h, loc=185, scale=8) #Probability that random number X is smaller than or equal to the height input
#Probability that height is within 1 standard deviation range 
b = scipy.stats.norm.cdf(up,loc=185, scale=8) 
c = scipy.stats.norm.cdf(low,loc=185, scale=8)
d = c - b

print(f"The probability that a random value is smaller than or equal to that value is {a}")
print(f"The probability of a height located within 1 standard deviation range is {d}")
```

```python
#SQ 4.6 
import numpy as np 
from numpy import *

#Input the function 
func = input(f"Function of x (Use np and x when necessary):")
a = float(input(f"Lower bound (a):"))
b = float(input(f"Upper bound (b):"))
n = int(input(f"How many random values are to be used in the summation?:"))

#Define the Monte Carlo integration 
def mci(func, a, b, n):
    rndm = np.random.uniform(a, b, n) #Sample n random points between a and b
    mci_2 = sum(eval(func, {"np": np, "x": rndm}))
    mci_1 = (b - a)/n
    mci_3 = mci_1 * mci_2
    return mci_3

ans = mci(func, a, b, n) 
print(f"The integral of the function using Monte Carlo integration is {ans}")
```

```python
#4.12  
import numpy as np
import matplotlib.pyplot as plt
import math

# Define angle in deg and convert to rad
theta_deg = 30
theta = np.deg2rad(theta_deg)

# Complex rotor
r = math.cos(theta) + 1j*math.sin(theta)

# Define the square
x = np.array([1, 3, 3, 1, 1])
y = np.array([1, 1, 3, 3, 1])

# Represent vertices as complex numbers
z = x + 1j*y

# Apply complex rotor 
z_rot = z * r

# Get the rotated x/y
x_rot = np.real(z_rot)
y_rot = np.imag(z_rot)

# Plot original and rotated square
plt.plot(x, y, 'b-', label='Original Square')
plt.plot(x_rot, y_rot, 'r--', label=f'Rotated {theta_deg}°')
plt.xlabel('x')
plt.ylabel('y')
plt.gca().set_aspect(1)
plt.title('Square rotation by complex rotor')
plt.grid(True, alpha=0.4)
plt.legend()
plt.show()
```

```python
#4.14
import numpy as np

#Create the function
def funcfilter(func, seq):
    seqar = np.asarray(seq) #Supports all sequences/single numbers by making them into an array, but not copying the array if an array is given
    filt = seqar > 0 #Filters out negative numbers and 0
    filtarr = seqar[filt] #Creates an array with the filtered numbers
    funcsol = func(filtarr) #Applies all in filtarr to the function
    return funcsol

#Test
sq = np.array([-100, -10, 0, 10, 100])
result = funcfilter(np.log10, sq)
print("Result log10:", result)    
```

```python
#SQ 4.21
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import matplotlib

# Workaround for older NumPy versions (added due to an error)
if not hasattr(np, 'int'):
    np.int = int

# Check matplotlib version and adjust parameters if needed (added due to an error)
matplotlib_version = matplotlib.__version__

# Seed for reproducibility
np.random.seed(0)

# Generate data
N = 100000
X = np.random.normal(loc=0, scale=1, size=N)
Y = np.random.normal(loc=2, scale=1, size=N)

# 1D Histogram + PDF Plot
fig, ax = plt.subplots(figsize=(7, 6))  # Creates a new figure and one axis 
bins = 30

# Set range and bins manually for control of axis
hist_range = (-6, 8)  # Sets the range of the histogram 
edges = np.linspace(hist_range[0], hist_range[1], bins + 1)  # Creates bins + 1 evenly spaced values (the bin edges, thus why we need 31)

# Plot histogram for X (blue)
counts_X, bin_edges_X, patches_X = ax.hist(  # Plots the histogram for X
    X,
    bins=edges,  # Uses our bin edges
    density=True,  # Normalizes the histogram to have a total area of 1
    alpha=0.6,
    color="royalblue",
    label="X",
    edgecolor="black",
)

# Plot histogram for Y (red)
counts_Y, bin_edges_Y, patches_Y = ax.hist(
    Y,
    bins=edges,
    density=True,
    alpha=0.6,
    color="salmon",
    label="Y",
    edgecolor="black",
)

# Ensure histogram bar outlines are solid and of standard thickness
for p in patches_X + patches_Y:  # For both histograms 
    p.set_linewidth(1.2)
    p.set_linestyle("solid")

# Plot PDF for X and Y
x_pdf = np.linspace(hist_range[0], hist_range[1], 500)  # Creates evenly spaced values for the x pdf
ax.plot(  # Plots the pdf for X
    x_pdf,
    norm.pdf(x_pdf, 0, 1),  # Uses the normal distribution function to plot the pdf for X
    "k:",  # Black dotted line
    lw=3,
    label="X(PDF)",
)
ax.plot(
    x_pdf,
    norm.pdf(x_pdf, 2, 1),
    "k--",
    lw=3,
    label="Y(PDF)",
)

# Configure axes limits and ticks
ax.set_xlim(hist_range)  # Sets the x axis to the range of the histogram
ax.set_ylim(0, 0.40)  # Sets the y axis to the range of the histogram
ax.set_xticks(np.arange(hist_range[0], hist_range[1] + 1, 1))  # Sets the tick interval for the x axis
ax.set_yticks(np.arange(0, 0.41, 0.05))  # Sets the tick interval for the y axis

# Labels, title, grid, legend
ax.set_xlabel("Values")
ax.set_ylabel("Probability Density")
ax.set_title("1D Histogram + PDF")
ax.grid(axis="y", linestyle="--", alpha=0.6)
ax.legend()

# Annotations for distributions
ax.annotate(
    r"$\mathcal{N}(0,1)$", 
    xy=(-1.1, 0.16),  # The coordinates of the annotation
    xytext=(-3.0, 0.19),
    arrowprops=dict(arrowstyle="->", color="blue", lw=2.2),  # The arrow style, color, and width
    color="blue",
    fontsize=12,
    ha="center",  # Horizontal alignment
)
ax.annotate(
    r"$\mathcal{N}(2,1)$",
    xy=(2.4, 0.16),
    xytext=(4.2, 0.19),
    arrowprops=dict(arrowstyle="->", color="red", lw=2.2),
    color="red",
    fontsize=12,
    ha="center",
)

plt.tight_layout()  # Adjusts the layout to prevent overlap

# 2D Histogram + Contour Plot
fig2, ax2 = plt.subplots(figsize=(8, 6))  # Creates a new figure and one axis 

# Create 2D histogram manually using numpy to avoid matplotlib version issues
hist, xedges, yedges = np.histogram2d(X, Y, bins=30)

# Create a meshgrid for plotting
x_centers = 0.5 * (xedges[:-1] + xedges[1:])
y_centers = 0.5 * (yedges[:-1] + yedges[1:])
X_mesh, Y_mesh = np.meshgrid(x_centers, y_centers, indexing="ij")

# Plot the 2D histogram as an image
img = ax2.pcolormesh(xedges, yedges, hist.T, cmap="Blues", shading='auto')
cb = plt.colorbar(img, ax=ax2)  # Adds a color bar to the axis
cb.set_label('Counts')

# Compute the contour levels
Z_flat = hist.ravel()  # Flattens the array
positive = Z_flat[Z_flat > 0]  # Finds the positive values
if positive.size == 0:  # If no positive values
    contour_levels = []  # Set the contour levels to an empty list
else:
    contour_levels = np.quantile(positive, [0.4, 0.7, 0.9])  # Finds the contour levels

# Red contour lines 
if len(contour_levels) > 0:
    CS = ax2.contour(  # Plots the contour lines
        X_mesh,
        Y_mesh,
        hist,
        levels=contour_levels,
        colors="red",
        linewidths=2,
    )

    # Label inner/middle/outer as 1σ/2σ/3σ
    sigma_text = ["1σ", "2σ", "3σ"]
    # Make sure we have enough labels for the levels
    labels_to_use = sigma_text[:len(CS.levels)]
    fmt = {level: label for level, label in zip(CS.levels, labels_to_use)}  # Creates a dictionary of the contour levels and the labels
    ax2.clabel(CS, CS.levels, inline=True, fmt=fmt, fontsize=10, colors="red")  # Labels the contour lines

ax2.set_title("2D Histogram + Contour")
ax2.set_xlabel("X")
ax2.set_ylabel("Y")
ax2.grid(True, linestyle="--", alpha=0.6)
ax2.set_aspect("equal", adjustable="box")  # Sets the aspect ratio to be equal

plt.tight_layout()
plt.show()
```
