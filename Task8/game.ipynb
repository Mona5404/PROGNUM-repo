```python
# Basic arcade shooter
#Bullet hell game? 

# Imports
import arcade
import random

# Constants
SCREEN_WIDTH = 800
SCREEN_HEIGHT = 600
SCREEN_TITLE = "The Greatest PROGNUM Game Ever Created"
SCALING = 2.0

#Open m101.fits and get its data to use
from astropy.visualization import ZScaleInterval
from astropy.io import fits
from matplotlib import pyplot as plt

#Currently shows axes and a lot of white space; can remove white space? maybe remove axes?
data = fits.getdata("190515_Li_00000080.fits") 
zscale = ZScaleInterval()
plt.imshow(zscale(data), cmap="twilight")
plt.savefig("bkg.png")

#Template
class PROGNUMGAME(arcade.Window):

    """Space Shooter side scroller game

    Player starts on the left, enemies appear on the right

    Player can move anywhere, but not off screen

    Enemies fly to the left at variable speed

    Collisions end the game

    """


    def __init__(self, width, height, title):

        """Initialize the game

        """

        super().__init__(width, height, title)

        #Background image
        self.background = "bkg.png"        
        
        # Set up the empty sprite lists

        self.enemies_list = arcade.SpriteList()

        self.clouds_list = arcade.SpriteList()

        self.all_sprites = arcade.SpriteList()
```
