# Energy and frequency

With Planck's constant $h$, we write the relation between energy and frequency as:

$$E = hf$$

$$E = hc/\lambda$$
*or, in the other notation,*
$$E = \frac{hc}{\lambda}$$

$$F = G\frac{m_1m_2}{r^2}$$ 

$$\sin(2x) = 2\sin(x)\cos(x) = \frac{2\tan(x)}{1+\tan^2(x)}$$

```python
#Import all of math
from math import *

#Set x as 30 deg, and change it to radians
alpha_d = 30 
alpha_r = radians(alpha_d)
x = alpha_r

#Solve each equation using an example x, e.g. 30 deg
ans1 = sin(2*x)
ans2 = 2*sin(x)*cos(x)

#Print the answer
print(f"The first equation gives us {ans1}, the second {ans2}, therefore this holds true for {alpha_d} degrees")
```

```python
#2.12 (Session Quiz) 

from math import atan2, degrees

#Input values
x = float(input("Input an x value:"))
y = float(input("Input a y value:"))

#Calculate angle, change to degrees
ans_rad = atan2(y,x)
ans_deg = degrees(ans_rad)

#If statements; checking the angle to know what quadrant it is in 
if 0 <= ans_deg < 90:
    print (f"It is in quadrant 1")
elif 90 <= ans_deg < 180:
    print (f"It is in quadrant 2")
elif 180 <= ans_deg < 270:
    print (f"It is in quadrant 3")
else:
    print (f"It is in quadrant 4")

#Print the angle 
print (f"The angle in degrees is {ans_deg}")
```

```python
#2.14 (Session Quiz)

#Creating a list X with values
X = [1.36,  1.46,  1.41,  2.12,  0.42,  0.98,  0.45,  1.97, -0.24, 2.18]

#Adding/removing values
X.append(0.51)
X.remove(2.12)

print(f"The value at index 6 is {X[6]}")

#Find the index of a value, assign it to a variable
a = X.index(1.46)
print(f"The index value of 1.46 is {a}")

#Slice, we want to start at index n=2, only including even index numbers except for the last even index number 
sl = slice(2,-2,2)
X_slice = X[sl]
print(f"The slice gives {X_slice}")

#Sum up the sliced list
sum = sum(X_slice) 
print(f"The sum of this sliced list is {sum}")
```

```python
#2.15 (Session Quiz)

#Create lists
night1 = [1,3,2,4,0,3,7,5,2,9,2,4,3]
night2 = [0,0,2,4,2,4,3,2,1,1,1,2,0]

#Make them into sets
sn1 = set(night1)
sn2 = set(night2)

#Find common values
totn = set.intersection(sn1, sn2)
print(f"The common values between the two nights are {totn}")

#Find if 5 is in the intersection list 
if 5 in totn: 
    print(f"The value 5 is an element of the intersection list")
else:
    print(f"The value 5 is not an element of the intersection list") 

#Find if 5 is in the set night 1 list 
if 5 in sn1: 
    print(f"The value 5 is an element of the set of night 1 list")
else:
    print(f"The value 5 is not an element of the set of night 1 list") 
```
