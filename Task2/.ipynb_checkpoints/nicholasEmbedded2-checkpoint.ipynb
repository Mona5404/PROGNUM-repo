2.16
Compiler - Languages in which the source code is translated into machine code (what the computer can directly read and execute). Compiled programs run faster oftentimes, but the compiling requires time, skill and patience, where the machine instructions are generated.    
Interpreter - Languages in which the code, or high level instructions, are translated through a medium into an intermediate form, which is then executed. Interpreters, unlike compilers, can directly execute high level programs     
Python binding - (Pre)compiled code that can be directly called upon and used in Python

```python
#Q2.19
#Initial part of the question; defining lists X and Y
import math
X = range(-10, 10)
Y = []
for x in X:
      y = math.exp(-(x/5)**2)
      Y.append(round(y,3))

#Creating them as lists (and printing them)
i = list(X)
j = list(Y)
print("The list X is ", i)
print("The list Y is ", j)

n = 2 #Defining n for the index
print(i[1:-1:n]) #all even index numbers in list i
print(j[1:-1:n]) #all even index numbers in list j
```

```python
#Q2.20

night1 = [1,3,2,4,0,3,7,5,2,9,2,4,3]
night2 = [0,0,2,4,2,4,3,2,1,1,1,2,0]

#Converting into sets
n1 = set(night1)
n2 = set(night2)
#Finding the intersection
inters = set.intersection(n1, n2)
print("Intersection of the 2 nights is", inters)

#Finding if 5 is in the intersection list
print(5 in inters, "for 5 in intersection list")

#Finding if 5 is in night 1 list
print(5 in n1, "for 5 in night 1 list")
```
