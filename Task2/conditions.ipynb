## Conditions

Essential in a programming language is that you can execute code depending on some condition.  
A condition is a statement like ``5 > 3``. Python evaluates these statements to either ``True`` or ``False``. 

```python
condition = 5 > 3          # Bigger than
print(condition)

condition = 3 < 0          # Smaller than
print(condition)

condition = 4 == 4         # Equal to
print(condition)

condition = 3 >= 3         # Equal to or bigger than
print(condition)

condition = 0 < 5 < 100    # Between
print(condition)
```

The last one is used to test if a values is in a range between to given values. It is an example of a
**chained comparison**.  

## Indentation
<font color='red'>IMPORTANT: The role of Indentation</font>

But how do we use these conditions? We introduce the ``if`` then ``else`` construction here. These statements use **indentation** to distinguish code that must be executed if the condition is true and code that must be executed if the condition is false. Indented code form a *block*. A block always starts with a colon``:``. The syntax is then:

    if condition 1:
        statement 1
        statement 2
        .....
    elif condition 2:
        statement 3
        statement 4
        .....
    elif
        statement 5
        statement 6
        .....
    ....
    else:
        statement 7
        statement 8
        .....
        
First a simple example:        

```python
M = 250
condition = M > 1000
if condition == True:
    print("M is more than 1000")
else:
    print("M is not more than 1000")
```

**Important:** But there is no need to check if the condition is equal to True, because the condition itself is already ``True`` or ``False``, so we rewrite the code above into:

```python
M = 250
condition = M > 1000
if condition:
    print("M is more than 1000")
else:
    print("M is not more than 1000")
```

Usually you use the condition only once, then you can write the code even shorter as in:

```python
M = 250
if M > 1000:
    print("M is more than 1000")
else:
    print("M is not more than 1000")
```

A condition can be composed from multiple comparisons. The most familiar operators are ``and`` and ``or``. Details about Boolean operators can be found in:
https://docs.python.org/3/library/stdtypes.html#boolean-operations-and-or-not

Here we give some examples that you can study. You should be able to predict the output of each line.

```python
redshift= 1
z = 2.31
print(redshift > z)
print(redshift < z)
print(redshift <= z)
print(redshift >= z)
print(redshift == z)
print(redshift != z)
print("Composed")
# Composed conditions
print( redshift > z or redshift <= 3 )
print( redshift < z and redshift >= 1 )
print( not (z == 0) )
print( not z == 0 )
print( z != 0 )
print( -z > 0 )
print(-(z > 0) )
print( -1*True )
print( 3*True )
print( 3*False )
print( not redshift < z and redshift >= 1 )
print( not (redshift < z and redshift >= 1) )
print( not True and False )
print( not (True and False) )
```

## Special constructions that result in ``True`` or ``False``

In one of the examples you see that you can multiply ``True`` and ``False`` with an integer. All values not equal to 0 can serve as a ``True`` and 0 can serve as a False. Empty strings, tuples and lists also evaluate to ``False``. 
In the cell below, we show some examples. We use the conversion function ``bool()`` to convert results to Booleans ``True`` or ``False``.

```python
s = ''
print("a=", bool(s))

s = 'hello'
print("s=", bool(s))

x = []
print("x=", bool(x))

x = [1,2,3]
print("y=", bool(x))

x = ()
print("z=", bool(x))

x = (1,2,3)
print("u=", bool(x))

x = None                            # None is explained in one of the next cells
print("v=", bool(x))

x = not None
print("w=", bool(x))

x = 0.0
print("k=", bool(x))

x = -12.2
print("m=", bool(x))
```

| Operator        | Remark                                      |

|logical not      | not True==False; not False==True            |
|logical and      | x > 0 and x < y                             |
|logical or       | x < 0 or x > 10                             |


## Special Python assignment using  ``None``

In many algorithms it is convenient if there is a way to check whether a variable has been given a value or not. Python provides a special value (in fact it is an object) for this. It is written as ``None``. 

One assigns ``None`` to a variable as follows: ``x = None`` and therefore we can test if this variable has value ``None`` with statement:

    if x == None:
       ....

But ``None`` is a fixed object and objects can be compared with operator ``is`` which is more efficient and handy to use when we are using arrays.

    if x is None:
       ....
