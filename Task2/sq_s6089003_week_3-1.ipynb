```python
#Session quiz 3.2
#Import
from scipy.constants import G
from math import *

#Lists
masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']

escv = {} #Dictionary for escape velocities

#Calculate vesc for each planet
for n, M, r in zip(names, masses, radii): #zip to pair corresponding elements from each list
    vesc = sqrt((2*G*M)/r)
    m2 = f"{M:.2e}" #Masses rounded to 2 d.p.
    print(f"The mass of {n} is {m2}") 
    escv[n] = vesc #Add the escape velocities to the dictionary with the names as keys

print(escv)
```

```python
#Session quiz 3.17
from math import * 
import numpy as np 

ar = input(f"Array:")
sp = ar.split(',') #Splits it and removes commas s.t. the user input is usable
numbers = []
for x in sp:
    numbers.append(float(x)) #Change into float, put in a list
arr = np.array(numbers) #Change list into array 

#Create the standard deviation function 
def stdev(arr):
    mu = arr.sum() / arr.size #Define mu
    x = (arr - mu)**2 #The difference (in the brackets)
    v = x.sum() / arr.size #Variance
    stdev = v ** 0.5
    return stdev

cool = stdev(arr) #Calling my function
boring = np.std(arr) #Comparing to numpy function

print(f"The manually coded stdev() gives {cool}, whereas the np.std() gives {boring}")

#One liner 
q = input(f"Do you want to input a new array for the one liner? Y/N").upper()
if q == "Y": 
    olar = input(f"Array (one liner):")
    olarr = np.array([float(x) for x in olar.split()]) #One liner input
else:
    olarr = arr

def stdevol(arr):
    stdevol = (((arr - arr.sum() / arr.size)**2).sum() / arr.size)**0.5 
    return stdevol

ol = stdevol(olarr)

if q == "Y":
    print(f"The one liner, with the new array, gives {ol}")
else:
    print(f"The one liner gives {ol}")
```
