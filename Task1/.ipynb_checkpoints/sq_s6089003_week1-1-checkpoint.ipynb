1. It lists all files with a number 0-9 in the file name and either c or o file extensions
2. The square brackets are used to indicate a list or a range of things (depending on what is written in between them) that is to replace the position it is in. Asterisks (*) indicate any number of those items (one or more) can replace it (in any order), while their absence indicates that it is only to be replaced with 1; files with 2 or more will not be included, e.g. in the list.
3. The difference between ls *[0-9].[co] and  ls * [0-9] *.co (ignore the spaces between the asterisks for the second, it is necessary to avoid it becoming italics) is that the first one gives options for the file name to contain numbers between 0 to 9 and either c or o **only** as the file extension, while the second still has the option for the file name to contain numbers between 0 to 9 but it has to have .co  as the file extension.
